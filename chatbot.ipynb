```python
from chatterbot import ChatBot
```

```python
from chatterbot.trainers import ListTrainer
from tkinter import *
```

```python
import pyttsx3 as pp #library for voice from bot
```

```python
engine=pp.init()
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[0].id)

def speak(word):
    engine.say(word)
    engine.runAndWait()

bot = ChatBot("Phone Specification")
convo = [
    "Hey",
    "Good Afternoon. Welcome Do you want know the specification of phones.",
    "Yes",
    """Please choose the phones from the list: 
    OnePlus 9R 
    Apple iPhone 11 
    Samsung Galaxy A22
    Samsung Galaxy A52
    Apple iPhone 12
    Galaxy Note 20
    Vivo iQoo Z3""",
    "OnePlus 9R",
    """RAM: 8 GB
    Processor: Qualcomm Snapdragon 870
    Rear Camera: 48 MP + 16 MP + 5 MP + 2 MP
    Front Camera: 16 MP
    Battery: 4500 mAh
    Display: 6.55 inches
    Price: Rs39,999""",
    "Apple iphone 11",
    """RAM: 4 GB
    Processor: Apple A13 Bionic
    Rear Camera: 12 MP + 12 MP
    Front Camera: 12 MP
    Battery: 3110 mAh
    Display: 6.1 inches
    Price:49,999""",
    "Vivo iQoo Z3",
    """RAM: 8 GB
    Processor: Qualcomm Snapdragon 870
    Rear Camera: 48 MP + 13 MP + 2 MP
    Front Camera: 16 MP
    Battery: 4400 mAh
    Display:6.62 inches
    Price: 31990""",
    "Samsung Galaxy A52",
    """RAM: 6 GB
    Processor: Qualcomm Snapdragon 720G
    Rear Camera: 64 MP + 12 MP + 5 MP + 5 MP
    Front Camera: 32 MP
    Battery: 4500 mAh
    Display: 6.5 inches
    Price:26,499""",
    "Apple iPhone 12",
    """RAM: 4 GB
    Processor: Apple A14 Bionic
    Rear Camera: 12 MP + 12 MP
    Front Camera: 12 MP
    Battery:2815 mAh
    Display: 6.1 inches
    Price:72,900""",
    "Galaxy Note 20",
    """RAM: 8 GB
    Processor: Samsung Exynos 9 Octa 990
    Rear Camera: 12 MP + 64 MP + 12 MP
    Front Camera: 10 MP
    Battery: 4300 mAh
    Display: 6.7 inches
    Price:73,999""",
    
]
trainer = ListTrainer(bot)

trainer.train(convo)
# answer = bot.get_response("What are the different models of phone?")
# print(answer)

# print("You can communicate with the bot now")
#
# while True:
#     query=input();
#     if query== 'exit':
#         break
#     answer=bot.get_response(query)
#     print("Bot : ",answer)


main = Tk()
main.geometry("550x600")
main.title("Phone Specification")

img= PhotoImage(file="bot.png")
photol = Label(main, image=img)

photol.pack(pady=5)

def ask_from_bot():
    query= textF.get()
    answer_from_bot = bot.get_response(query)
    msgs.insert(END, "you : " + query)
    msgs.insert(END, "bot : " + str(answer_from_bot))
    speak(answer_from_bot)
    textF.delete(0,END)
    msgs.yview(END)  #to show the last message either from bot or you without scrolling

frame=Frame(main)
sc=Scrollbar(frame)

msgs= Listbox(frame,width=80,height=15,yscrollcommand=sc.set)
sc.pack(side=RIGHT,fill=Y)

msgs.pack(side=LEFT, fill=BOTH, pady=10)

frame.pack()

#creation of field to put questions

textF= Entry(main, font=("Times", 20, "bold italic"))
textF.pack(fill=X,pady=10)

#creating button
btn = Button(main, text="Ask?",font=("Times", 20, "bold italic"),command=ask_from_bot)
btn.pack()

#when we press enter to ask bot
def enter_function(event):
    btn.invoke()

main.bind('<Return>', enter_function)


main.mainloop()
```
